# Setup the Project Environment for CircuitBox Data Lakehouse
1. Create external location- dea_cource_ext_dl_circuitbox
2. Creating Catalog - CircuitBox
3. Create Schemas
    - landing
    - lakehouse
4. Create Vloumne- operational_date

### 1. Create external location- dea_cource_ext_dl_circuitbox

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS dea_cource_ext_dl_circuitbox
    URL 'abfss://circuitbox@databricksprostoageacc.dfs.core.windows.net/'
    WITH (STORAGE CREDENTIAL `databricks-course-ext-ac`)
    COMMENT 'This is ext loc for CircuitBox catalog'
```

### 2. Creating Catalog - CircuitBox

```python
%sql
CREATE CATALOG IF NOT EXISTS catalog_circuitbox
    MANAGED LOCATION 'abfss://circuitbox@databricksprostoageacc.dfs.core.windows.net/'
    COMMENT 'Catalog for CircuitBox'
```

### 3. Create Schemas
    1. landing
    2. lakehouse

```python
%sql
USE CATALOG catalog_circuitbox;

CREATE SCHEMA IF NOT EXISTS landing
    MANAGED LOCATION 'abfss://circuitbox@databricksprostoageacc.dfs.core.windows.net/landing';

CREATE SCHEMA IF NOT EXISTS lakebouse
    MANAGED LOCATION 'abfss://circuitbox@databricksprostoageacc.dfs.core.windows.net/lakebouse';
```

```python
%sql
SHOW SCHEMAS;
```

### 4. Create Volume- operational_date

```python
%sql
USE CATALOG catalog_circuitbox;
USE SCHEMA landing;

CREATE EXTERNAL VOLUME IF NOT EXISTS operational_data
    LOCATION 'abfss://circuitbox@databricksprostoageacc.dfs.core.windows.net/landing/operational_data';
```
